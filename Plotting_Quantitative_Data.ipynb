```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```
MPGdf = pd.read_csv("https://static.junilearning.com/python_data_science/mpg.csv")
display(MPGdf)
```

<h3>Plotting Distributions</h3>

```
sns.histplot(MPGdf["weight"], bins=30)
```

```
sns.histplot(MPGdf["weight"], bins=100)
```

```
sns.histplot(MPGdf["mpg"], bins=45)
#In histograms, median is center of peak. Mean is on the right side of the median if it's right skew.
#For symmetric bell curves, median and mean are in the middle.
```

```
sns.kdeplot(MPGdf['mpg'], color="pink")
plt.title("Distribution Plot of Car's MPG")
```

<h3>Plotting Relationships</h3>

```
sns.scatterplot(x = "weight", y = "mpg", data = MPGdf)
#Explanatory variable (x axis) and response variable (y axis). For example - health is y, age is x. 
```

```
#What does truncate do? False is preferred, goes throughout whole graph.

sns.regplot(x = "weight", y = "mpg", data = MPGdf, truncate=False, line_kws={"color":"orange"})


#Orange line is "linear regression line". "Line of best fit". #Below is inverse relationship and the two vars are negatively correlated. The opposite is a direct relationship or positively correlated.
```

U.S Unemployment Rate 1990-2015

```
unemploymentDF = pd.read_csv("https://static.junilearning.com/python_data_science/unemployment.csv")
display(unemploymentDF)
sns.lineplot(x="Year", y = "Rate", data=unemploymentDF)
```

```
maineOxford = unemploymentDF[(unemploymentDF["County"] == "Oxford County") & (unemploymentDF["State"] == "Maine")]
sns.lineplot(x="Year", y="Rate", data=maineOxford)
plt.show()
```

```
NYNassau = unemploymentDF[(unemploymentDF["County"] == "Nassau County") & (unemploymentDF["State"] == "New York")]
sns.lineplot(x="Year", y="Rate", data=NYNassau)
plt.show()
```

```
NYNassau = unemploymentDF[(unemploymentDF["County"] == "Nassau County") & (unemploymentDF["State"] == "New York")]
sns.lineplot(x="Year", y="Rate", data=NYNassau)

maineOxford = unemploymentDF[(unemploymentDF["County"] == "Oxford County") & (unemploymentDF["State"] == "Maine")]
sns.lineplot(x="Year", y="Rate", data=maineOxford)

plt.title("Unemployment Rate Over Time")
plt.ylabel("Unemployment Rate")
plt.xlabel("Time (Year)")
plt.xlim(2000,2015)

plt.grid(True)
plt.savefig("prettygraph.png")

plt.show()
```

<h3>Basic Customizations</h3>

```
# Basic Customization

plt.title("Weight vs MPG")
plt.xlabel("Weight (in Pounds)")
plt.ylabel("MPG")
plt.xlim(0, 5000)
plt.ylim(0, 50)
plt.grid(axis="both")

sns.regplot(x = "weight", y = "mpg", data = MPGdf, truncate=False, line_kws={"color":"orange"})
plt.show()
plt.savefig("mpgRegplot.png")
```
